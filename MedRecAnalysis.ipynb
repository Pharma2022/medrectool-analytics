```python
import pandas as pd
```

```python

url ='https://raw.githubusercontent.com/Pharma2022/medrectool-analytics/main/medrec.csv'
# df= pd.read_csv("medrec.csv",index_col=None) remove #if running locally from this line and delete next line
df= pd.read_csv(url,index_col=None)

df.rename(columns={'Total Number of medications  (incl OTC +ACUTE)':'Total Number of medications'},inplace=True)
df.drop(columns=['Modified By','ID','Hospital No','Number of Sources used'],inplace=True)
df_main =df.groupby('MedRecTool or Manual').mean(numeric_only=True).apply(lambda x: round(x,1)).drop(columns='Errors Found')
df_main['Average time per item (min)']=df_main[ 'Time taken to TYPE drug history on EPMA (MINUTES)']/df_main['Total Number of medications']
df_main['Average time per item (min)']=df_main['Average time per item (min)'].apply(lambda x: round(x,2))
df_count=df.groupby('MedRecTool or Manual').count().loc[:,['Ward']].rename(columns={'Ward':'No Patients'})
df_errors=df.groupby('MedRecTool or Manual').sum(numeric_only=True).loc[:,['Errors Found']]
df_main_stats= pd.concat([df_count,df_main,df_errors,], axis=1)
df_main_stats
```

```python
# MMTs vs Pharmacists
# df= pd.read_csv("medrec.csv",index_col=None) remove # if running locally from this line and delete next line
df= pd.read_csv(url,index_col=None)
df.rename(columns={'Total Number of medications  (incl OTC +ACUTE)':'Total Number of medications'},inplace=True)
df.drop(columns=['Modified By','ID','Hospital No','Number of Sources used'],inplace=True)
df_main =df.groupby(['MedRecTool or Manual','MMT/PHARMACIST']).mean(numeric_only=True).apply(lambda x: round(x,1)).drop(columns='Errors Found')
df_main['Average Drug History Time']=df.groupby(['MedRecTool or Manual','MMT/PHARMACIST']).mean(numeric_only=True).loc[:,['Time taken to TYPE drug history on EPMA (MINUTES)']].apply(lambda x:round(x,2))
df_main['Average time per item (min)']=df_main[ 'Time taken to TYPE drug history on EPMA (MINUTES)']/df_main['Total Number of medications']
df_main['Average time per item (min)']=df_main['Average time per item (min)'].apply(lambda x: round(x,2))
df_count=df.groupby(['MedRecTool or Manual','MMT/PHARMACIST']).count().loc[:,['Ward']].rename(columns={'Ward':'No Patients'})
df_errors=df.groupby(['MedRecTool or Manual','MMT/PHARMACIST']).sum(numeric_only=True).loc[:,['Errors Found']]
df_main_stats= pd.concat([df_count,df_main,df_errors,], axis=1).apply(lambda x: round(x,2))
df_main_stats.unstack().to_csv('mmtvspharmacist.csv')
df_main_stats
```
